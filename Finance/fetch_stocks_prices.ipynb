## Stocks from Yahoo Finance

This Notebook shows how to pull stock prices using Yahoo finance API and pandas data library.   
You need to install Pandas_data reader, run `pip install pandas_datareader` if you don't have it already, and ofcourse you need pandas and matplotlib.   


First import libraries


```python
import pandas as pd
import pandas_datareader as web
import datetime as dt
```

Three simple steps as follows:  
 1. Save required tickers in a list.   
 2. Specifiy start and end dates as `datetime`  
 3. Pull the data

```python
#
ticker = ['TSLA','MSFT']
start = dt.date(2020,1,1)
end = dt.date.today()
```

```python
df = web.DataReader(ticker, 'yahoo', start, end)
```

```python
df.head()
```

```python
df[['Adj Close']].plot(subplots=True, figsize=(10,8))
df[['Adj Close']].rolling(14).mean().plot(subplots=True, figsize=(10,8))
```

```python
import mplfinance as plf

spy = web.DataReader('SPY', 'yahoo', start, end)
```

```python
plf.plot(spy, volume=True, style='mike', type='candle', mav=[14,52], figsize=(10,8))
```

```python
import numpy as np
```

```python
date_range = pd.date_range(start, end, periods=500)
```

```python
x = np.random.randn(500).cumsum()
```

```python
series = pd.Series(data= x, index=date_range, name='X')
```

```python
series.plot()
```
